```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style='darkgrid', palette='Set2')
plt.rcParams.update({'figure.facecolor': '#0a0a1f', 'axes.facecolor': '#0a0a1f',
                     'text.color': 'white', 'axes.labelcolor': 'white',
                     'xtick.color': 'white', 'ytick.color': 'white'})

print('Libraries loaded ✓')
```

## 1. Load Datasets

```python
app    = pd.read_csv('../data/application_record.csv')
credit = pd.read_csv('../data/credit_record.csv')

print(f'application_record shape : {app.shape}')
print(f'credit_record shape      : {credit.shape}')
```

## 3. Univariate Analysis — Countplots

```python
cat_cols = ['NAME_INCOME_TYPE', 'NAME_EDUCATION_TYPE',
            'NAME_FAMILY_STATUS', 'NAME_HOUSING_TYPE']

fig, axes = plt.subplots(2, 2, figsize=(16, 10))
axes = axes.flatten()

for i, col in enumerate(cat_cols):
    order = app[col].value_counts().index
    sns.countplot(x=col, data=app, palette='Set2', ax=axes[i], order=order)
    axes[i].set_title(f'Distribution of {col}', fontweight='bold')
    axes[i].set_xlabel('')
    axes[i].tick_params(axis='x', rotation=30)

plt.suptitle('Categorical Feature Distributions', fontsize=16, fontweight='bold', y=1.01)
plt.tight_layout()
plt.show()
```

```python
# OCCUPATION_TYPE separately (has nulls)
fig, ax = plt.subplots(figsize=(12, 5))
order = app['OCCUPATION_TYPE'].value_counts().index
sns.countplot(x='OCCUPATION_TYPE', data=app, palette='Set1', ax=ax, order=order)
ax.set_title('Occupation Type Distribution (note: ~30.5% missing)', fontweight='bold')
ax.tick_params(axis='x', rotation=40)
plt.tight_layout()
plt.show()
```

## 4. Multivariate Analysis — Correlation Heatmap

```python
numeric_df = app.select_dtypes(include=[np.number])
print('Numeric columns:', list(numeric_df.columns))

fig, ax = plt.subplots(figsize=(12, 8))
mask = np.triu(np.ones_like(numeric_df.corr(), dtype=bool))
sns.heatmap(
    numeric_df.corr(), annot=True, fmt='.2f',
    cmap='coolwarm', mask=mask,
    linewidths=0.5, ax=ax,
    cbar_kws={'shrink': 0.8}
)
ax.set_title('Feature Correlation Heatmap', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```
